```python
import json
import csv
import pandas as pd
from transformers import AutoTokenizer, MarianMTModel
```

```python
df = pd.read_csv('common_japanese_words.csv', encoding='utf-8')
df.head()
```

```python
subdf = df[['japanese', 'pronounciation', 'translation']]
subdf.head()
```

```python
words = subdf.to_dict(orient='records')
print(f"Total: {len(words)}")
```

```python
with open('japaneseWords.json', 'w', encoding='utf-8') as f:
    json.dump(words, f, ensure_ascii=False, indent=2)
```

```python
with open('japaneseWords.json', 'r', encoding='utf-8') as f:
    words = json.load(f)

print(f"Total: {len(words)}")
```

```python
model_name = f"Helsinki-NLP/opus-mt-en-es"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
for word in words: 
    text = word["translation"]
    batch = tokenizer([text], return_tensors="pt")

    generated_ids = model.generate(**batch)
    translated_text = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    word["spanish"] = translated_text
    print(word)

    
with open('japaneseWords.json', 'w', encoding='utf-8') as f:
    json.dump(words, f, ensure_ascii=False, indent=2)
```
