```python
import json
```

```python
with open('englishWords.json', 'r', encoding='utf-8') as f:
    words = json.load(f)

print(f"Total: {len(words)}")
```

```python
from transformers import AutoTokenizer, MarianMTModel
```

```python
model_name = f"Helsinki-NLP/opus-mt-en-es"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
for word in words: 
    text = word["word"]
    batch = tokenizer([text], return_tensors="pt")

    generated_ids = model.generate(**batch)
    translated_text = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    word["spanish"] = translated_text
    print("Translated:", translated_text, "for:", word)

    
with open('englishWords.json', 'w', encoding='utf-8') as f:
    json.dump(words, f, ensure_ascii=False, indent=2)
```
